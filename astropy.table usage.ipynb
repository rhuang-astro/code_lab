# 使用Tbale 转换成 array

```python
import numpy as np
import astropy
from astropy.table import Table
```

## start from a numpy array

```python
np.arange(256)
```

## array to Table

```python
s=Table([np.arange(256)],names='0')
s1=Table([np.arange(256)],names='1')
s
```

## merge table

```python
ss = astropy.table.hstack([s,s1])
ss
```

## Convert table to structured array

```python
sss = np.array(ss)
sss
```

## Convert structured array to regular NumPy array

```python
sss.view(np.float64).reshape(sss.shape + (-1,))
```

## save table data as fits

```python
ss.write('test10.fits', overwrite=True)
```
